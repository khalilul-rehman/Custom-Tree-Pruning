```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

df = pd.read_csv("custom_tree_experiments_results.csv")

# Derived metrics
df["nodes_reduction"] = df.nodes_before - df.nodes_after
df["nodes_reduction_pct"] = df.nodes_reduction / df.nodes_before

df["NRMSE_degradation"] = (
    df.NRMSE_Test_Affine_after
    - df.NRMSE_Test_Affine_before
)

df["NRMSE_vs_tree"] = (
    df.NRMSE_Test_Affine_after
    - df.NRMSE_Test_Tree
)
```

# SEABORN DASHBOARDS

```python
g = sns.FacetGrid(
    df,
    col="system_name",
    row="cost_function",
    hue="max_depth",
    sharey=True,
    height=4
)

g.map_dataframe(
    sns.lineplot,
    x="alpha",
    y="NRMSE_Test_Affine_after",
    marker="o"
)

# g.set(xscale="log")
g.add_legend()
g.fig.suptitle("Alpha vs Test Error After Pruning", y=1.02)
plt.show()
```

```python
g = sns.relplot(
    data=df,
    x="nodes_after",
    y="NRMSE_Test_Affine_after",
    hue="alpha",
    col="cost_function",
    row="system_name",
    palette="viridis",
    height=4
)

g.set_titles("{row_name} | {col_name}")
g.fig.suptitle("ComplexityAccuracy Tradeoff", y=1.03)
plt.show()
```

```python
# Compression vs Degradation
sns.lmplot(
    data=df,
    x="nodes_reduction_pct",
    y="NRMSE_degradation",
    hue="cost_function",
    col="system_name",
    height=4,
    aspect=1.2
)

plt.suptitle("Compression vs Error Increase", y=1.02)
plt.show()
```

```python
# Heatmap: Alpha  Depth
pivot = df.pivot_table(
    index="alpha",
    columns="max_depth",
    values="NRMSE_Test_Affine_after",
    aggfunc="mean"
)

plt.figure(figsize=(8,6))
sns.heatmap(pivot, annot=True, fmt=".4f", cmap="viridis")
plt.title("Avg Test Error After Pruning")
plt.show()
```

# STATISTICAL TESTS

```python
# Paired: before vs after.
# If p < 0.05  pruning significantly changes accuracy
before = df["NRMSE_Test_Affine_before"]
after = df["NRMSE_Test_Affine_after"]

t_stat, p_val = stats.ttest_rel(before, after)

print("Paired t-test: Before vs After")
print("t =", t_stat)
print("p =", p_val)
```

```python
# Compare basic vs average after pruning.
basic = df[df.cost_function == "basic"]["NRMSE_Test_Affine_after"]
avg = df[df.cost_function == "average"]["NRMSE_Test_Affine_after"]

u_stat, p_val = stats.mannwhitneyu(basic, avg)

print("MannWhitney U test: basic vs average")
print("p =", p_val)
```

```python
# Correlation: nodes vs error
r, p = stats.spearmanr(
    df.nodes_after,
    df.NRMSE_Test_Affine_after
)

print("Spearman correlation:")
print("rho =", r)
print("p =", p)
```

```python
# Effect of Alpha (ANOVA)
stats.f_oneway(
    *[
        df[df.alpha == a]["NRMSE_Test_Affine_after"]
        for a in df.alpha.unique()
    ]
)
```
